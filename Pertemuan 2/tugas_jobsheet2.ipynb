```python
import pandas as pd
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import OneHotEncoder, StandardScaler

df = pd.read_csv('wbc.csv') # load dataset
df.head()
```

```python
df.info()
df.isnull().sum()
```

## Soal 1 Slice Data
Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

Jawab :
id dan unnamed:32 tidak akan digunakan karena bukan merupakan atribut yang mempengaruhi klasifikasi kanker.

```python
# Disini saya membuang kolom 'id' dan 'Unnamed: 32'
df = df.drop(columns=['id', 'Unnamed: 32'], errors='ignore')
df.head()
```

# Soal 2 Encoding
Lakukan proses encoding pada kolom "diagnosis".

```python
# 2. Lakukan proses encoding pada kolom "diagnosis".
encoder = OneHotEncoder(sparse_output=False)

diagnosis_encoded = encoder.fit_transform(df[['diagnosis']])
diagnosis_encoded_df = pd.DataFrame(diagnosis_encoded, columns=encoder.get_feature_names_out(['diagnosis']))
df_encoded = pd.concat([df, diagnosis_encoded_df], axis=1)
df_encoded = df_encoded.drop(columns=['diagnosis'])

df_encoded.head()
```

# Soal 3 Standarization
Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.


Jawab : Standarisasi mengecualikan diagnosis_B dan diagnosis_M menggunakan diferensiasi karena merupakan data label

```python
# Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

std = StandardScaler()

numeric_columns = df_encoded.columns.difference(['diagnosis_B', 'diagnosis_M'])
df_encoded[numeric_columns] = std.fit_transform(df_encoded[numeric_columns])

df_encoded.head()
```

# Soal 4 Stratified Split Data
 Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio
80:20.

```python
#  Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

x = df_encoded.drop(columns=['diagnosis_B', 'diagnosis_M'])
y = df_encoded[['diagnosis_B', 'diagnosis_M']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=0)

print(len(x_test))
print(len(x_train))
print(len(y_test))
print(len(y_train))
```
